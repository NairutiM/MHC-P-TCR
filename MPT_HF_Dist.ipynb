```python
import os
import math
import numpy as np
import pandas as pd 
import os
import shutil
os.chdir (r'/Users/nairutimehta/Desktop/MHCPTCR/Conf Files_3Res')
```

```python
def category(dist):
    """
    The category of distance.
    """
    if dist < 0.5:
        cate = "<0.5"
    elif dist >= 0.5 and dist < 1:
        cate = "[0.5, 1]"
    elif dist >= 1 and dist < 1.5:
        cate = "[1, 1.5]"
    elif dist >= 1.5 and dist < 3:
        cate = "[1.5, 3]"
    elif dist >= 3 and dist < 5:
        cate = "[3, 5]"
    else:
        cate = ">=5"
    return cate

def distance_calculate2(pdb_name, lines):
    """
    Calculate the distance between atoms.
    """
    carbons = ["CA ","CB ","CG ","CD ","CE ","CZ ","CH ","CD1","CD2","CG1","CG2","CE1","CE2","CE3","CZ1","CZ2","CH1","CH2"]
    atom_name = []
    chain_name = []
    res_name = []
    res_num = []
    x_coord = []
    y_coord =[]
    z_coord =[]
    for line in lines:
        if line[13:16] in carbons:
            atom_name.append(str(line[13:16]))
            chain_name.append(str(line[21:22]))
            res_name.append(str(line[17:20]))
            res_num.append(int(line[23:26]))
            x_coord.append(float(line[31:38]))
            y_coord.append(float(line[39:46]))
            z_coord.append(float(line[47:54]))

    output = []
    for i in range(0, len(x_coord), 1):
        for j in range(i+1, len(x_coord), 1):
            if chain_name [i] == "A" and chain_name[j] == "C":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = math.sqrt(xx+yy+zz)
                    cate = category(dist)
                    output.append([ atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
            if chain_name [i] == "B" and chain_name[j] == "C":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = math.sqrt(xx+yy+zz)
                    cate = category(dist)
                    output.append([ atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
            if chain_name [i] == "C" and chain_name[j] == "D":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = math.sqrt(xx+yy+zz)
                    cate = category(dist)
                    output.append([ atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
            if chain_name [i] == "C" and chain_name[j] == "E":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = math.sqrt(xx+yy+zz)
                # Added column: distance category. You can filter and count.
                    cate = category(dist)
                    output.append([ atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
            
    output = np.array(output)
    return output
```

```python
Peptide_dir = r'/Users/nairutimehta/Desktop/MHCPTCR/Conf Files_3Res' 
files = os.listdir(Peptide_dir)
for file in files:
    if file.endswith('.pdb'):
        src = r'/Users/nairutimehta/Desktop/MHCPTCR/Conf Files_3Res/' + file
        pdb_name = src
        input_file = src 
        with open(input_file,'r') as pdb_file:
            lines = pdb_file.readlines()
        file_name3 = file + "_HF_dist.xlsx"
        res = distance_calculate2(pdb_name, lines)
        columns = [ "atom name i", "chain name i", "res name i", "res num i", "atom name j", "chain name j", "res name j", "res num j", "distance", "distance category"] 
        pd.DataFrame(res, columns=columns).to_excel(file_name3, index=None)
```
